```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_clients = 1000

clients = pd.DataFrame({
    "Client_ID": [f"C{i:04d}" for i in range(1, n_clients + 1)],
    "Region": np.random.choice(["North", "South", "East", "West"], n_clients),
    "Annual_Sales": np.random.normal(500000, 150000, n_clients).astype(int),
    "Purchase_Frequency": np.random.randint(5, 80, n_clients),
    "Average_Order_Value": np.random.normal(12000, 4000, n_clients).astype(int),
    "Discount_Rate": np.random.uniform(0.01, 0.25, n_clients).round(2),
    "Product_Diversity": np.random.randint(1, 20, n_clients),
    "Payment_Delay_Days": np.random.randint(0, 60, n_clients),
    "Years_As_Client": np.random.randint(1, 15, n_clients),
})

clients["Customer_Loyalty_Index"] = (
    (clients["Purchase_Frequency"] * clients["Years_As_Client"]) / 
    (1 + clients["Payment_Delay_Days"])
).round(2)


clients.to_csv("data/synthetic_clients.csv", index=False)
print("✅ Synthetic dataset created.")
```
